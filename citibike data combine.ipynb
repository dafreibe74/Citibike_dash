```python
import pandas as pd
```

```python
#grab CSVs
df_1 = pd.read_csv('citibike_6_19.csv')
df_2 = pd.read_csv('citibike_7_19.csv')
df_3 = pd.read_csv('citibike_8_19.csv')
df_4 = pd.read_csv('citibike_9_19.csv')
df_5 = pd.read_csv('citibike_6_20.csv')
df_6 = pd.read_csv('citibike_7_20.csv')
df_7 = pd.read_csv('citibike_8_20.csv')
df_8 = pd.read_csv('citibike_9_20.csv')
```

```python

# Combine the DataFrames vertically
combined_df = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8], ignore_index=True)
```

```python
#check dataframe
combined_df.head()
```

```python
# drop unneeded columns

combined_df.drop(columns=['bikeid', 'stoptime', 'start station id', 'end station id'], inplace=True)
```

```python
#check dataframe
combined_df.head()
```

```python
# update gender values

combined_df["gender"].replace({1: 'Male', 2: 'Female', 0: 'Unknown'}, inplace=True)

combined_df.head()
```

```python
# create "trip year" field for analysis

combined_df['Trip Year'] = combined_df.starttime.str.split("-", expand=True)[0]
combined_df.head()
```

```python
#caluclate Rider age
combined_df['Trip Year'] = combined_df['Trip Year'].astype(int)
combined_df['birth year'] = combined_df['birth year'].astype(int)

combined_df["Rider Age"] = combined_df["Trip Year"]-combined_df["birth year"]
combined_df.head()
```

```python
# export CSV

combined_df.to_csv('combined_data.csv', index=False)
```
